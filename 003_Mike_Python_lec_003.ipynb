###Seaborn Library:

```python
%reset -f
import numpy as np ; 
import seaborn as sns  ; 
import matplotlib as plt ; 

import pandas as pd ; 
%whos
```

```python
# Creating Some Random data:

n = 200 ; 
D = np.zeros((200 , 2))

# Populate the elements of data matrix D:
D[: , 0] = np.linspace(0 , 10 , n) + np.random.randn(n)
D[: , 1] = D[: , 0]**2 + np.random.randn(n) * 10 ; 


# Using Seaborn:
sns.jointplot(D[:,0] , D[: , 1])
# plt.show() 
```

```python
# Create a Pandas DataFrame from DataMatrix D:

df = pd.DataFrame(data=D , columns = ['var1' , 'var2'])

sns.jointplot(df.columns[0] , df.columns[1] , data = df)
# plt.show()
```

```python
# KDE : Kernal Density Estimator:
sns.jointplot(df.columns[0] , df.columns[1] , data = df , kind = 'kde' , color = 'purple')
```

```python
# Creating Some Random data:

n = 200 ; 
D = np.zeros((200 , 2))

# Populate the elements of data matrix D:
D[: , 0] = np.linspace(0 , 10 , n) + np.random.randn(n)
D[: , 1] = D[: , 0]**2 + np.random.randn(n) * 10 ; 
df = pd.DataFrame(data = D , columns=['var1' , 'var2'])
print(df)

# Regression PLot:
fig , ax = plt.subplots(figsize = (10 , 8))
sns.regplot(df.columns[0] , df.columns[1] , data = df , color = 'g')
plt.title('My First Regression Plot' , fontsize = 18)
plt.show()
```

###Visualizing Matrices as Images:

```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
%whos
```

```python
m = 3 ; 
n = 5 ; 

fig , ax = plt.subplots(figsize = (12,8) )
M = np.random.randint(9 , size = (m,n))
print(M)

plt.imshow(M , cmap = 'jet') ; 
for i in range(m):
  for j in range(n):
    plt.text(j , i , str(M[i,j]) , horizontalalignment = 'center' , fontsize = 30 , verticalalignment = 'center' ) ; 

plt.colorbar()
plt.show()
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

gray_img = rgb2gray(img) ; 
plt.figure( figsize = (20 , 12) ) ; 
plt.imshow(gray_img , cmap = 'gray') ; # , vmin = 50 , vmax = 200 )
plt.colorbar() ; 
plt.xticks([]) ; 
plt.yticks([]) ; 
```

```python
# Exercise:
#   1) Create a Hilbert Matrix (10by10)
#   H(i,j) = 1/(i+j-1)
#   2) Visualise the hilbert matrix using seaborn!
#   3) Try using sns.heatmap() for visualization:
 
%reset -f
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%whos
```

```python
n = 10 ; 
H = np.zeros((n , n))

for i in range(1 , n+1):
  for j in range(1 , n+1):
    H[i-1 , j-1] = 1 / (i + j - 1) ; 

sns.heatmap(H , cmap = 'jet' , vmin = 0 , vmax = 0.3)
```

###Export Plot In Low and High Dimension:


```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')

%whos
```

```python
x   = np.linspace(0.5 , 5 , 15)
y1  = np.log(x) ; 
y2  = 2-np.sqrt(x) ; 

fig = plt.figure(1 , figsize = (10 , 8) )
plt.plot(x , y1 , 'bv-' , lw = 2 , ms = 12 , mfc = 'w' , label = 'log')
plt.plot(x , y2 , 'rs-' , lw = 2 , ms = 12 , mfc = 'w' , label = 'sqrt')

plt.legend()
plt.savefig('Test.pdf')
plt.show()
```
